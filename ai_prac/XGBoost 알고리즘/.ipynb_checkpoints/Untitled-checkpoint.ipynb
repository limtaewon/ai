### XGBoost를 이용한 clasification

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from xgboost import XGBClassifier
from xgboost import plot_importance
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.info()
```

```python
iris.head()
```

```python
cols = list(iris.columns)
```

```python
col_x = cols[:4]
```

```python
col_y = cols[-1]
```

```python
iris_train,iris_test = train_test_split(iris,test_size=0.2,random_state=123)
```

```python
iris_train.shape
```

```python
iris_test.shape
```

```python
model = XGBClassifier()
```

```python
model.fit(X=iris_train[col_x],y=iris_train[col_y])
model
```

```python
fscore = model.get_booster().get_fscore()
fscore
# y에 영향을 끼치는 요소
```

```python
plot_importance(fscore)
```

```python
y_pred = model.predict(iris_test[col_x])
y_true = iris_test[col_y]
```

```python
acc = accuracy_score(y_pred,y_true)
acc
```

```python
con_mat = confusion_matrix(y_true,y_pred)
con_mat
```

```python
report = classification_report(y_true,y_pred)
print(report)
```

```python
fig = plt.figure(figsize=(12,5))
```

### XGBoost를 이용한 Regression

```python
from xgboost import XGBRegressor
from xgboost import plot_importance

from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
boston = load_boston()
```

```python
X=boston.data
y=boston.target
```

```python
col_names = boston.feature_names
```

```python
X.shape
```

```python
y.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
plot_importance(model)
plt.yticks(range(13),col_names)
plt.show()
```

```python
y_pred = model.predict(X_test)
y_true = y_test
```

```python
mae = mean_absolute_error(y_true,y_pred)
mse = mean_squared_error(y_true,y_pred)
mae
```

```python
mse
```

### Regressor 실습 1

```python
from xgboost import XGBRegressor
from xgboost import plot_importance,plot_tree

from sklearn.datasets import fetch_california_housing
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
housing = fetch_california_housing()
X = housing.data
y = housing.target
cols_name = housing.feature_names
```

```python
X.shape
```

```python
y.shape
```

```python
import numpy as np
```

```python
y.mean()
```

```python
y = np.log(y) # 특이값이나 편향을 제거하기 위함
y
```

```python
y.mean()
```

```python
X_df = pd.DataFrame(X,columns = cols_name)
X_df
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_df,y,test_size=0.25)
```

```python
model = XGBRegressor(n_estimators=400) #n_estimators는 트리의 수를 의미
model.fit(X_train,y_train)
```

```python
plot_importance(model)
```

```python
y_pred = model.predict(X_test)
y_true = y_test
```

```python
mse = mean_squared_error(y_true,y_pred)
mse
```

```python
fig = plt.figure(figsize=(12,4))
chart = fig.add_subplot(1,1,1)
chart.plot(y_true[:100],marker='o',color='blue',label='real')
chart.plot(y_pred[:100],marker='^',color='red',label='predict')
plt.xlabel('index')
plt.ylabel('value')
plt.title('real and predict compare')
plt.legend()
plt.show
```

### Classifier 실습

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from xgboost import plot_importance
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
import time
```

```python
freeze = pd.read_csv('freeze_dataset.csv',encoding = 'ms949')
```

```python
freeze.shape
```

```python
freeze.info()
```

```python
freeze.columns = freeze.columns.str.replace(' ','_')
freeze.info()
```

```python
freeze.동파유무.value_counts()
```

```python
checktime = time.time()
```

```python
X_train,X_test = train_test_split(freeze,test_size=0.4)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
cols = list(freeze.columns)
col_x = cols[1:]
col_y = cols[0]
```

```python
model = XGBClassifier(n_estimator = 400)
model.fit(X_train[col_x],X_train[col_y])
model
```

```python
fscore = model.get_booster().get_fscore()
fscore
len(fscore)
sorted(fscore)
```

```python
names = sorted(fscore,key=fscore.get,reverse=True)
names
```

```python
score = [fscore[key] for key in names]
score
```

```python
fscore_df = pd.DataFrame({'names':names,'score':score})
fscore_df
```

```python
new_fscore_df = fscore_df.set_index('names')
new_fscore_df.iloc[:10,:].plot(kind='bar')
```

```python
y_pred = model.predict(X_test[col_x])
y_true = X_test[col_y]
acc = accuracy_score(y_true,y_pred)
acc
```

```python
curr_time = time.time() - checktime

```

```python
con_mat = confusion_matrix(y_true,y_pred)
```

```python
report = classification_report(y_true,y_pred)
```

```python
print('='*40)
print('>> 소요시간 :', curr_time)
print(con_mat)
print(report)
```
