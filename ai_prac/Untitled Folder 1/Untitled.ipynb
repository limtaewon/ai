```python
from selenium import webdriver
import time
 
# 크롬 웹브라우저 실행
path = "C:\\chromedriver_win32\\chromedriver.exe"
```

```python
driver = webdriver.Chrome(path)
url_list = []
content_list = ""
text = "제주도애월"
```

```python
for i in range(1, 10):  # 1~2페이지까지의 블로그 내용을 읽어옴
    url = 'https://section.blog.naver.com/Search/Post.naver?pageNo='+ str(i) + '&rangeType=ALL&orderBy=sim&keyword=' + text
    driver.get(url)
    time.sleep(0.5)
 
    for j in range(1, 8): # 각 블로그 주소 저장
        titles = driver.find_element_by_xpath('//*[@id="content"]/section/div[2]/div['+str(j)+']/div/div[1]/div[1]/a[1]')
        title = titles.get_attribute('href')
        url_list.append(title)
 
print("url 수집 끝, 해당 url 데이터 크롤링")
 
for url in url_list: # 수집한 url 만큼 반복
    driver.get(url) # 해당 url로 이동
 
    driver.switch_to.frame('mainFrame')
    overlays = ".se-component.se-text.se-l-default" # 내용 크롤링
    contents = driver.find_elements_by_css_selector(overlays)
 
    for content in contents:
        content_list = content_list + content.text # content_list 라는 값에 + 하면서 점점 누적
```
