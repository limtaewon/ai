```python
user_data = pd.DataFrame(data=[],columns=['INDEX','사용자 ID','관광지이름','사용자별 리뷰점수','리뷰내용','사진유무','작성일'])
tour_data = pd.DataFrame(data=[],columns=['INDEX','관광지이름','관광지 소개글','리뷰갯수','총 리뷰점수','긍정적 리뷰 개수','주소','사진갯수','좋아요갯수'])
```

```python
관광지이름 = []
관광지_소개글 = []
총_리뷰점수 = []
리뷰갯수 = []
긍정적_리뷰개수 = []
주소 = []
사진갯수 = []
좋아요갯수 = []

사용자_ID = []
관광지이름2 = []
사용자별_리뷰점수 = []
리뷰내용 = []
사진유무 = []
작성일 = []
```

```python
site = pd.read_excel('site_url.xlsx')
site_url = list(site[0])
num = 0
```

```python
driver = webdriver.Chrome("C:\\chromedriver_win32\\chromedriver.exe")
```

```python
for i in range(1):
    driver.get(site_url[i])
    time.sleep(1)
    
    destination= driver.find_element_by_css_selector('#poi\.detail\.overview > div > h1 > div')
    관광지이름.append(destination.text.strip())
    total_review_score = driver.find_element_by_css_selector('#poi\.detail\.overview > div > div > div > div.ovh.burited_point.header > div.l.rating-review > span.gl-poi-detail-rating > span:nth-child(1)')
    총_리뷰점수.append(total_review_score.text.strip())
    try:
        like = driver.find_element_by_css_selector('#poi\.detail\.overview > div > div > div > div.ovh.burited_point.header > div.favourite-wrapper > div > div > span')
        좋아요갯수.append(like.text.strip())
    except:
        좋아요갯수.append('NULL')
    address = driver.find_element_by_css_selector('#poi\.detail\.overview > div > div > div > div.mt10.ovh.top-box-relative > div.gl-poi-detail_top-time-map.gl-poi-detail_top_time-map2 > div.POITopInfo-jljzy3-13.iVlEOq.address-map.flex.clickable > div.address > div > span.field')
    주소.append(address.text.strip())
    
    driver.find_element_by_css_selector("#poi\.detail\.overview > div > div > div > div.ovh.burited_point.header > div.l.rating-review > span.gl-poi-detail_review > div").click()
    
    review_count = driver.find_element_by_css_selector('#reviews > div > div > div > div.ContentContainer-tlt00z-0.bddgEJ > div:nth-child(1) > div.SwitchContainer-x26rcx-0.fQxJGb > div.SwitchListContainer-x26rcx-2.bshVSp > div.switch-item.switch-item-all')
    리뷰갯수.append(review_count.text.strip())
    good_review = driver.find_element_by_css_selector('#reviews > div > div > div > div.ContentContainer-tlt00z-0.bddgEJ > div:nth-child(1) > div.SwitchContainer-x26rcx-0.fQxJGb > div.SwitchListContainer-x26rcx-2.bshVSp > div:nth-child(3)')
    긍정적_리뷰개수.append(good_review.text.strip())
    try:
        picture_count = driver.find_element_by_css_selector('#reviews > div > div > div > div.ContentContainer-tlt00z-0.bddgEJ > div:nth-child(1) > div.SwitchContainer-x26rcx-0.fQxJGb > div.SwitchListContainer-x26rcx-2.bshVSp > div:nth-child(5)')
        사진갯수.append(picture_count.text.strip())
    except:
        picture_count = driver.find_element_by_css_selector('#reviews > div > div > div > div.ContentContainer-tlt00z-0.bddgEJ > div:nth-child(1) > div.SwitchContainer-x26rcx-0.fQxJGb > div.SwitchListContainer-x26rcx-2.bshVSp > div:nth-child(4)')
        사진갯수.append(picture_count.text.strip())
    while True:
        for i in range(1,7):
            for_error = 1
            while_error = 1
            try:
                content = driver.find_element_by_css_selector('#reviews > div > div > div > div.ContentContainer-tlt00z-0.bddgEJ > div:nth-child(1)> div.bgff.ovh > ul > div:nth-child('+str(i)+')'+' > li')
                driver.execute_script("arguments[0].scrollIntoView(true);", content)
            except:
                for_error = 0
            if for_error==0:
                continue
            user_name = driver.find_element_by_css_selector('#reviews > div > div > div > div.ContentContainer-tlt00z-0.bddgEJ > div:nth-child(1) > div.bgff.ovh > ul > div:nth-child('+str(i)+')'+' > li > div.user.review-user-view > div > div')
            사용자_ID.append(user_name.text.strip())
            time.sleep(0.1)
            관광지이름2.append(관광지이름[len(관광지이름)-1])
            user_per_score = driver.find_element_by_css_selector('#reviews > div > div > div > div.ContentContainer-tlt00z-0.bddgEJ > div:nth-child(1) > div.bgff.ovh > ul > div:nth-child('+str(i)+')'+' > li > div.gl-poi-detail_comment-content > div:nth-child(1) > span:nth-child(1)')
            사용자별_리뷰점수.append(user_per_score.text.strip())
            time.sleep(0.1)
            try:
                driver.find_element_by_css_selector('#reviews > div > div > div > div.ContentContainer-tlt00z-0.bddgEJ > div:nth-child(1) > div.bgff.ovh > ul > div:nth-child('+str(i)+')'+' > li > div.gl-poi-detail_comment-content > div.read_more_btn').click()
            except:
                num = 0
            review = driver.find_element_by_css_selector('#reviews > div > div > div > div.ContentContainer-tlt00z-0.bddgEJ > div:nth-child(1) > div.bgff.ovh > ul > div:nth-child('+str(i)+')'+' > li > div.gl-poi-detail_comment-content > a > p')
            리뷰내용.append(review.text.strip())
            time.sleep(0.1)
            day = driver.find_element_by_css_selector('#reviews > div > div > div > div.ContentContainer-tlt00z-0.bddgEJ > div:nth-child(1) > div.bgff.ovh > ul > div:nth-child('+str(i)+')'+' > li > div.gl-poi-detail_comment-content > div.ovh.thumb-box > span > span')
            작성일.append(day.text.strip())
            time.sleep(0.1)
        time.sleep(1)    
        try:
            driver.find_element_by_css_selector('#reviews > div > div > div > div.ContentContainer-tlt00z-0.bddgEJ > div:nth-child(1) > div.bgff.ovh > div > div > div > button.btn-next.disabled').click()
            while_error = 0
        except:
            driver.find_element_by_css_selector('#reviews > div > div > div > div.ContentContainer-tlt00z-0.bddgEJ > div:nth-child(1) > div.bgff.ovh > div > div > div > button.btn-next').click()
        if while_error == 0:
            break
        time.sleep(0.5)
        
        
    content = driver.find_element_by_css_selector('#details > div > div > div > div > div')
    driver.execute_script("arguments[0].scrollIntoView(true);", content)
    
    try:
        dt_introduce = driver.find_element_by_xpath('//*[@id="details"]/div/div/div/div/div/div/div/text()')
        관광지_소개글.append(dt_introduce.text.strip())
    except:
        관광지_소개글.append('NULL')
    
    
```
