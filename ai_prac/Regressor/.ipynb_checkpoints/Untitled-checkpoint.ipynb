##### 군집-정답이 없는 문제를 해결하기 위해 사용하는 알고리즘. 정답을 모르는 데이터 안에서 숨겨진 구조를 찾는 것. 
##### K-평균_k-means: 군집의 하위 분야. 구현이 쉽고 다른 군집 알고리즘에 비해 높은 계산 효율성을 갖는다. 학계나 산업현장 모두 사용한다. 프로타타입 기반 군집에 속한다
##### 프로토타입: 연속적인 특성에서는 비슷한 데이터 포인트의 센트로이드_평균이거나 범주형 특성에서는 메도이드_가장 자주 등장하는 포인트가 된다. 
##### 프로토타입 군집: 각 클러스터가 하나의 프로터타입으로 표현된다는 의미
##### K-MEANS는 원형 클러스터를 구분하는데 뛰어나지만 KNN과 마찬가지로 몇개의 클러스터를 만들것인지 직접 지정해줘야 한다. 
##### 우리의 목표는 특성의 유사도에 기초하여 데이터들을 그룹으로 모으는 것이다. 
##### K-MEANS알고리즘 4단계 요약 1. 데이터 포인트에서 랜덤하게 k개의 센트로이드를 초기 클러스터 중심으로 선택 2. 각 데이터를 가장 가까운 센트로이드에 할당 3. 할당된 샘플들의 중심으로 센트로이드를 이동 4. 클러스터 할당이 변하지 않거나, 사용자가 지정한 허용 오차나, 최대 반복횟수에 도달할때 까지 2,3과정을 반복한다
##### (유사도 측정)각 data를 가장 가까운 센트로이드에 할당할때마다 센트로이드는 이동 -> 센트로이드가 변화할때마다 오차 제곱합_sse를 반복적으로 계산하면서 변화에 대한 허용 오차값이 일정 수준내로 들어오면 더 이상 클러스터가 변하지 않고 최적화가 완료되었다는 것->점들간의 단위와 변동폭의 값 차이가 많이 나면 왜곡의 발생 확률이 증가. -> 단위를 일치시키는 표준화 과정 진행
##### K-means는 클러스터가 중첩되지 않고, 계층적이지 않고, 클러스터당 하나 이상의 데이터를 갖고 있다 가정하에 돌아간다 -> 큰 리스크를 가져온다 => k-means++ 사용
##### K-means++: 초기 클러스터 센트로이드를 똑똑하게 할당하는 기법. 초기 센트로이드를 멀리 떨어지도록 위치시키는 것.
##### 비지도 학습은 클래스 레이블이 없으므로 군집을 이루기의 최종 답을 모른다. -> 지도학습의 성능 평가 기준 적용 불가. -> 군집 품질을 평가하는 알고리즘 자체의 지표를 사용 => 군집 성능을 평가하기 위해 클러스터 내 오차 제곱합_SSE 사용->사이킷런 사용
##### Elbow: k를 다양한 값으로 할당 뒤 왜곡 값을 비교한다. 최적인 클러스터 개수k를 추정 가능하다. K의 증가->센트로이드의 증가->데이터들이 샌트로이드에 더 가까워 지므로 sse_왜곡값 감소. 왜곡이 빠르게 증가하는 지점 k를 찾는 것. 
##### 실루엣 분석: 군집 분석을 평가하는 또다른 방법. 클러스터 내 데이터들이 얼마나 조밀하게 모여있는지를 측정하는 그래프 도구. K-means외에도 다른 군집 알고리즘에 적용 가능.
##### 1. 하나의 임의의 데이터(x^i)와 동일한 클러스터내의 모든 다른 데이터 포인트 사이의 거리를 평균하여 클러스터 응집력(a^i)을 계산 2. 앞서 선정한 데이터와 가장 가ᄁᆞ운 클러스터의 모든 샘플간 평균 거리로 최근접 클러스터의 클러스터 분리도C를 계산 3. 클러스터 응집력과 분리도 사이의 차를 둘 중 큰값으로 나눠 실루엣 계수(s^i) 계산 -> 응집력과 분리도가 같다면 실루엣 계수는 0, 응집력보다 분리도가 크면 이상적인 실루엣 계수가 1에 가깝다. 
##### 분리도: 데이터가 다른 클러스터와 얼마나 다른지
##### 응집력: 클러스터 내 다른 샘플과 얼마나 비슷한지 ->응집력이 작을수록 클러스터 내 다른 데이터들과 비슷하다. 분리도와 응집력이 같다면 두 클러스터는 중첩되어 있다는 의미이다.  ==> 군집분석은 클래스 레이블이 없는 데이터를 특정 군집으로 묶을 때 활용하고, k-means알고리즘은 학계와 산업현장을 가리지 않고 활약, k-means의 무작위성을 보완하기 위해 k-means++도 널리 사용.
##### 계층 군집: 군집분석의 하위분야인 모델. 덴드로그램 사용. 클러스터의 수를 미리 지정하지 않아도 된다. 병합 계층 군집&분할 계층 군집
##### 덴도로그램: 의미 있는 분류 체계를 만들어 군집 결과 이해하는데 도움. 히트맵과 자주 사용
##### 히트맵: 이미지 위 혹은 측면에 데이터를 열분포 형태로 표현하는 시각화 기법, 데이터 행렬을 색으로 표현 가능, 각 데이터셋을 효과적으로 요약 가능
##### 분할계층 군집: 전체 데이터를 포함하는 하나의 클러스터에서 시작해 클러스터 속 데이터가 하나가 남을 때까지 반복적으로 클러스터를 나눈다.
##### 병합계층 군집: 하나의 데이터에서 시작해 하나의 클러스터에 모일 때까지 가까운 클러스터 병합. 단일 연결, 완전연결, 평균연결, 와드 연결
##### 1. 단일 연결: 클러스터 쌍에서 가장 가까운_비슷한 데이터 간의 거리를 계산, 거리가 가장 가까운 클러스터 끼리 병합 2. 완전 연결: 클러스터 상에서 가장 멀리 있는_비슷하지 않은 데이터를 찾아 거리를 구하고 가장 가까운 클러스터 병합 3. 평균연결: 두 클러스터에 있는 모든 샘플 사이의 평균 거리가 가장 적은 클러스터 쌍을 합침 4. 와드 연결: SSE(두 군집을 합쳐졌을 때의 제곱 오차합)가 가장 작게 증가하는 두 클러스터를 합치는 방식
##### 완전연결의 동작 방식 1. 모든 데이터의 거리행렬을 계산한다. 1. 모든 데이터 포인트를 단일 클러스터로 표현한다. 2. 가장 비슷하지 않은 즉, 멀리 떨어진 데이터 간 거리에 기초하여 가장 가까운 두 클러스터를 하나로 합쳐준다. 3. 유사도 행렬 업데이트 4. 하나의 클러스터가 남을 때 까지 2~4 반복

## K-mean 실습1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
df = pd.DataFrame(columns=['x','y'])
j = 0
for i in range(12):
    df.loc[i] = [i+3,1+j]
    if (i+3)%2 == 0:
        j=j+1
```

```python
df
```

```python
sns.lmplot('x','y',data=df,fit_reg=False,scatter_kws={'s':200})
plt.title('kmean')
plt.xlabel('x')
plt.ylabel('y')
plt.show
```

```python
data_point = df.values
```

```python
data_point
```

```python
kmeans = KMeans(n_clusters=3).fit(data_point)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
df['cluster_id'] = kmeans.labels_
```

```python
df
```

```python
sns.lmplot('x','y',data=df,fit_reg=False,scatter_kws={'s':150},hue ='cluster_id')
```
