```python
cols = train.columns
cols_na = cols[3:-1]
```

```python
train.hour_bef_temperature.fillna(train.hour_bef_temperature.mean(),inplace = True)
train.hour_bef_precipitation.fillna(train.hour_bef_precipitation.mean(),inplace = True)
```

### id :  고유 id,  hour : 시간,  temperature : 기온,  precipitation : 비가오면 1 안오면 0, windspeed : 풍속,  humidity : 습도,  visibility : 가시성,  ozone : 오존,  pm10 : 미세먼지  , pm25 : 초미세먼지,  count : 시간에 따른 따릉이 대여 수

```python
train_test_data = [train,test]
```

```python
def change(feature):
    train[feature].fillna(train[feature].mean(),inplace = True)
```

```python
def change2(feature):
    test[feature].fillna(test[feature].mean(),inplace = True)
```

```python
for x in cols_na:
    change(x)
    change2(x)
```

```python
col_x = cols[1:-1]
col_y = cols[-1]
```

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier()
model.fit(train[col_x],train[col_y])
model
```

```python
y_pred = model.predict(test[col_x])
```

```python
id = test['id'].values
id
```

```python
df = pd.DataFrame({
    'id' : id,
    'count':y_pred
})
df
```
