#### 좋은 군집 : 그룹의 크기가 작고 그룹끼리의 거리가 먼 군집
#### silhouette conefficient (실루엣계수) : 1에가까울수록 최적화
#### Davies Bouldin Index  :  값이 작을수록 최적화

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head(10)
```

```python
df.info()
```

```python
del df['CustomerID']
```

```python
mapping = {'Male':0,'Female':1}
mapping
```

```python
df['Genre'] = df['Genre'].map(mapping)
df.head()
```

```python
df.rename(columns={'Genre':'Gender'},inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
sns.countplot('Gender',data=df)
```

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
pca = PCA(n_components = 2)
reduced_df = pca.fit_transform(df)
reduced_df
```

```python
n_df = pd.DataFrame(columns=['x','y'],data = reduced_df)
n_df
```

```python
fig = plt.figure(figsize=(12,6))
sns.lmplot('x','y',data=n_df,fit_reg=False,scatter_kws={'s':200})
# PCA과정을 거친 
plt.show
```

```python
from sklearn.cluster import KMeans
distortions = []
for k in range(2,20):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(reduced_df)
    distortions.append(kmeans.inertia_) # inertia는 응집도를 나타냄
    
fig = plt.figure(figsize=(10,5))
plt.plot(range(2,20),distortions)
plt.grid(True) #격자표시
plt.title('Elbow curve')
plt.show
```

```python
km = KMeans(n_clusters=5,init = 'k-means++')
cluster = km.fit(n_df)
cluster_id = pd.DataFrame(cluster.labels_)
fig = plt.figure(figsize=(12,6))
d1 = pd.concat([n_df,cluster_id],axis = 1)
d1.columns = [0,1,'cluster']

sns.scatterplot(d1[0],d1[1],hue=d1['cluster'],legend ='full')
sns.scatterplot(km.cluster_centers_[:,0],km.cluster_centers_[:,1],label='centroid')
plt.legend()
plt.show()
```

```python
from sklearn import metrics
from sklearn.metrics.cluster import silhouette_score
metrics.silhouette_score(d1.iloc[:,:-1],d1['cluster'])
```

```python
metrics.davies_bouldin_score(d1.iloc[:,:-1],d1['cluster'])
```

```python
pca = PCA(n_components=2)
reduced_iris = pca.fit_transform(iris)
reduced_iris
```

```python
distributor=[]
for k in range(2,10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(reduced_iris)
    distributor.append(kmeans.inertia_)
    
fig = plt.figure(figsize=(12,6))
plt.plot(range(2,10),distributor)
plt.grid(True)
plt.show()
```

```python
reduced_iris = pd.DataFrame(reduced_iris)

km = KMeans(n_clusters=3)
cluster = km.fit(reduced_iris)
cluster_id = pd.DataFrame(cluster.labels_)

d1 = pd.concat([reduced_iris,cluster_id],axis=1)
d1.columns = [0,1,'cluster']

sns.scatterplot(d1[0],d1[1],hue=d1['cluster'])
sns.scatterplot(km.cluster_centers_[:,0],km.cluster_centers_[:,1],label='centroid')
plt.legend()
plt.show()
```
