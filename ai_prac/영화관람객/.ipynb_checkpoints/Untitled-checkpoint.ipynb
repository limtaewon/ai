```python
train_test_data = [train,test]
```

```python
for data in train_test_data:
    data['distributor'] = data.distributor.str.replace('(주)','')
```

```python
for data in train_test_data:
    data['distributor'] = [re.sub('[^0-9a-zA-Z가-힣]','',x)for x in data.distributor]
```

```python
distributor = np.array(train['distributor'].tolist())
distributor = np.unique(distributor)
distributor.shape
```

```python
for x in distributor:
    print(x)
```

```python
def dis_mapping(data):
    if 'CJ' in data or 'CGV' in data:
        return 'CJ'
    elif '쇼박스' in data:
        return '쇼박스'
    elif 'SK' in data:
        return 'SK'
    elif '더빅쳐' in data:
        return '더빅쳐'
    elif '마운틴' in data:
        return '마운틴'
    elif '스폰지' in data:
        return '스폰지'
    elif '싸이더스' in data:
        return '싸이더스'
    else:
        return data
```

```python
for data in train_test_data:
    data['distributor'] = data.distributor.apply(dis_mapping)
```

```python
train.groupby('genre').box_off_num.mean().sort_values()
```

```python
for data in train_test_data:
    data['genre'] = data.genre.map({'뮤지컬':1,'다큐멘터리':2,'서스펜스':3,'애니메이션':4,'멜로/로맨스':5,'미스터리':6,'공포':7,'드라마':8,'코미디':9,'SF':10,'액션':11,'느와르':12})
```

```python
dis_nm_rank = train.groupby('distributor').box_off_num.median().reset_index(name='rank').sort_values(by='rank')
dis_nm_rank
```

```python
dis_nm_rank['rank'] = [i+1 for i in range(dis_nm_rank.shape[0])]
dis_nm_rank
```

```python
dir_nm_rank = train.groupby('director').box_off_num.mean().reset_index(name='dir_rank').sort_values(by='dir_rank')
dir_nm_rank
```

```python
dir_nm_rank['dir_rank'] = [i+1 for i in range(dir_nm_rank.shape[0])]
dir_nm_rank
```

```python
train = pd.merge(train,dis_nm_rank,how='left')
```

```python
train
```

```python
train = pd.merge(train,dir_nm_rank,how='left')
```

```python
train
```

```python
test = pd.merge(train,dis_nm_rank,how='left')
test = pd.merge(train,dir_nm_rank,how='left')
```

```python
train.fillna(0,inplace=True)
test.fillna(0,inplace=True)
```

```python
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold
from ngboost import NGBRegressor
```

```python
X = train[['rank','dir_rank','time','num_staff','num_actor','screening_rat']]
y = np.log1p(train.box_off_num)
```

```python
X=pd.get_dummies(columns = ['screening_rat'],data=X)
```

```python
X['num_actor'] = np.log1p(X['num_actor'])
```

```python
target = test[['rank','dir_rank','time','num_staff','num_actor','screening_rat']]
```

```python
target = pd.get_dummies(columns = ['screening_rat'],data=target)
```

```python
target['num_actor'] = np.log1p(target['num_actor'])
```

```python
kf = KFold(n_splits = 10, shuffle = True, random_state = 42)
```

```python
gbm = GradientBoostingRegressor(random_state = 42)
```

```python
rmse_list=[]
gb_pred = np.zeros((test.shape[0]))
for tr_idx,val_idx in kf.split(X,y):
    tr_x,tr_y = X.iloc[tr_idx],y.iloc[tr_idx]
    val_x,val_y = X.iloc[val_idx],y.iloc[val_idx]
    
    gbm.fit(tr_x,tr_y)
    
    pred = np.expm1([0 if x<0 else x for x in gbm.predict(val_x)])
    sub_pred = np.expm1([0 if x<0 else x for x in gbm.predict(target)])
    rmse = np.sqrt(mean_squared_error(val_y,pred))
    
    rmse_list.append(rmse)
    gb_pred += (sub_pred/10)
```

```python
np.mean(rmse_list)
```

```python
ngb = NGBRegressor(random_state = 518)
```

```python
cat = CatBoostRegressor(random_state = 518, silent = True)
```

```python
rmse_list = []
cat_pred = np.zeros((test.shape[0]))
for tr_idx, val_idx in kf.split(X, y) :
    tr_x, tr_y = X.iloc[tr_idx], y.iloc[tr_idx]
    val_x, val_y = X.iloc[val_idx], y.iloc[val_idx]
    
    cat.fit(tr_x, tr_y)
    
    pred = np.expm1([0 if x < 0 else x for x in cat.predict(val_x)])
    sub_pred = np.expm1([0 if x < 0 else x for x in cat.predict(target)])
    rmse = np.sqrt(mean_squared_error(val_y, pred))
    
    rmse_list.append(rmse)
    
    cat_pred += (sub_pred / 10)
```

```python
rmse_list = []
rf_pred = np.zeros((test.shape[0]))
for tr_idx, val_idx in kf.split(X, y) :
    tr_x, tr_y = X.iloc[tr_idx], y.iloc[tr_idx]
    val_x, val_y = X.iloc[val_idx], y.iloc[val_idx]
    
    rf.fit(tr_x, tr_y)
    
    pred = np.expm1([0 if x < 0 else x for x in rf.predict(val_x)])
    sub_pred = np.expm1([0 if x < 0 else x for x in rf.predict(target)])
    rmse = np.sqrt(mean_squared_error(val_y, pred))
    
    rmse_list.append(rmse)
    
    rf_pred += (sub_pred / 10)
```
